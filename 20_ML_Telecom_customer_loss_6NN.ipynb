## Телеком — задача проекта (оригинальное описание)

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

### Описание услуг

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

### Описание данных

Данные состоят из файлов, полученных из разных источников:

- `contract.csv` — информация о договоре;
- `personal.csv` — персональные данные клиента;
- `internet.csv` — информация об интернет-услугах;
- `phone.csv` — информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

````python
def first_info(dataframe, data_title='Неопознанный набор данных',
               head_length=6, bins=42, figsize=(15,12),
               info_show=True, describe_show=True, corr_show=True, hist_show=True):
    '''
    dataframe: датафрейм для анализа
    data_title: пользовательское название набора данных
    head_length: количество строк, отображаемых функцией при выводе датасета
    bins: для гистограммы - количество корзин, 42 by default
    figsize: для гистограммы - размер графика, кортеж (15,12) by default
    info_show: True, если нужно показывать результат метода DataFrame.info(), True by default 
    describe_show: True, если нужно показывать результат метода DataFrame.describe(), True by default 
    corr_show: True, если нужно показывать результат метода DataFrame.сorr(), True by default 
    hist_show: True, если нужно показывать гистограмму распределений значений DataFrame, True by default
    
    Функция  принимает на вход Pandas.DataFrame, Краткое название датафрейма (для вывода в логе выполнения функции) 
    и количество строк датафрейма, которое будет выводить функция в ходе работы. 
    Функция создана для упрощения первичного знакомства с данными
    Возвращает None
    '''
    try:
        print(f'Данные по датасету: {data_title}')
        display(dataframe.head(head_length))
        display(dataframe.tail(head_length))
        if info_show:
            dataframe.info()
        if describe_show:
            print()
            print(f'Описательная статистика датасета:')
            display(dataframe.describe())
        if corr_show:
            print('Корреляция значений признаков')
            display(pd.DataFrame(dataframe.corr()).round(2))
        if hist_show:    
            dataframe.hist(bins=bins, figsize=figsize);
        
    except:
        print(f'Ошибка в передаваемых аргументах\nПроверьте ввод!')
````

````python
def snake_case(string):
    '''Функция принмает строку, кодирует ее в змеиный регистр вовзращает предобработанную строку
    
    '''
    result = ''
    result = re.sub(r'(.)([A-Z])', r'\1_\2', string).lower()
    
    return result
````

````python
def best_corr(dataset, range_=(0.5, 1)):
    '''function displays best correlation coefficients for pairs of features in corr matrix
    return pd.DataFrame with pairs of features and their correlation coefficients
    '''
    corr_table = pd.DataFrame({'first_feature': [],
                               'second_feature': [],
                               'corr_coef': []})
    exist_list = []
    for col_name, data in dataset.items():
        exist_list.append(col_name)
    
        for second in data.items():
            corr_value = round(second[1], 2)
            if corr_value > range_[0] and corr_value <= range_[1]:
                if second[0] not in exist_list:
                    corr_table = corr_table.append({'first_feature': col_name,
                                                    'second_feature': second[0],
                                                    'corr_coef': corr_value},
                                                   ignore_index=True)  
                
    return corr_table.sort_values(by='corr_coef', ascending=False, ignore_index=True)
````

````python
def merge_plus(dataset_full, array_features, features_list):
    '''
    dataset_full: pandas.DataFrame (1)
    array_features: array-like (2)
    features_list: list-like
    
    Функция принимает полный датасет (1) и массив (2), 
    где были преобразованы значения признаков из списка features_list, 
    и возвращает полный датасет, в котором значения столбцов из списка features_list (1) были заменены на (2)
    Применяется для объединенного датасета с численными, категориальными и булевыми признаками,
    для трансформации отдельных признаков.
    
    Return pandas.Dataframe
    '''
    
    dataset_part = pd.DataFrame(array_features,
                                columns=dataset_full[features_list].columns)
    
    suffixes_tuple = ('_left', '')
    dup_features_after_merge = [i+suffixes_tuple[0] for i in features_list]

    dataset_part = pd.merge(dataset_full, dataset_part, 
                                     how ='left', left_index=True, right_index=True, 
                                     suffixes=suffixes_tuple)
    
    return dataset_part.drop(columns=dup_features_after_merge, inplace=False)
    
````

````python
def del_features(dataset, features_list):
    '''Local function
    The function delete features in features_list from dataset 
    and from next lists: cat_features, date_features, num_features, binary_features
    return dataset
    
    '''
    for _ in features_list:
        try:
            dataset.drop(columns=_, inplace=True)
        except:
            pass

        try:
            cat_features.remove(_)
        except ValueError:
            pass
    
        try:
            date_features.remove(_)
        except ValueError:
            pass
    
        try:
            num_features.remove(_)
        except ValueError:
            pass
    
        try:
            binary_features.remove(_)
        except ValueError:
            pass

    return dataset
````

 - имеющие корреляцию выше 0,8 с признаками входящими в список признаков наиболее коррелирующих с целевым  
 - имеющие низкую статистическую значимость корреляции
 - имеющие низкую корреляцию < 0.1 с целевым признаком
 
Безусловно сохраняются призанки

````python
# Определение признаков с низакой статистической значимостью коэффициента корреляции Фk

# Задаем нижний порог, значение квантиля, признаки с суммой статзначимости лежащие ниже будут удалены
min_quantile = 0.05

# 1 - создание списка списков с суммами статзначимостей коэффициентов корреляции с 
# остальными признаками для каждого признака
a_ = [[c_, significance_overview[c_].sum()] for c_ in significance_overview.columns] 

# Преобразование в Pandas DataFrame
a_ = pd.DataFrame(a_, columns=['feature_name', 'stat'])

# Фильтрация значения ниже квантиля min_quantile
feat_min_stat = a_[a_['stat'] < a_['stat'].quantile(min_quantile)].reset_index(drop=True)
````

Булевы функции - это случайные величины Бернулли, а дисперсия таких переменных определяется выражением $\mathrm{Var}[X] = p(1 - p)$ то есть в случае p = 80%: 0.8 * (1 - 0.8)

````python
sel = VarianceThreshold(threshold=(0.8 * (1 - 0.8)))
sel.fit_transform(ds_train[binary_features])
_ = pd.Series(binary_features)[(sel.get_support() != True)].to_list()

print('Удаляемые признаки с низкой дисперисей:\n', _)
````

#### Проверка методом SequentialFeatureSelection

Чтобы получить представление о важности признаков, примением следующие оценщики:

- RidgeClassifierCV
- LogisticRegressionCV
- LGBMClassifier
- RandomForestClassifier

На каждой итерации процесса SequentialFeatureSelection мы выбираем лучший признак для добавления к выбранным признакам на основе оценки перекрестной проверки. 

То есть мы начинаем с 0 признаков и выбираем лучший единственный признак с наивысшим баллом - метод forward.
Обратный метод - начиная со всех признаков, последовательно удалять неважные - метод backward.

Процедура повторяется до тех пор, пока мы не достигнем желаемого количества выбранных признаков.

Для задач оценки методом SequentialFeatureSelection используя модели LogisticRegressionCV и RidgeClassifierCV предобработку признаков и сами модели используем с предыдущего этапа (SelectFromModel)

````python
def sfs_research(model_, feature_names_, X_, y_, on_backward=True):
    '''
    '''
    sfs_forward = SequentialFeatureSelector(
        model_, n_features_to_select=4, direction="forward").fit(X_, y_)
    
    if on_backward:
        sfs_backward = SequentialFeatureSelector(
            model_, n_features_to_select=4, direction="backward").fit(X_, y_)

    return feature_names_[sfs_forward.get_support()], feature_names_[sfs_backward.get_support()]
````

````python
%%time
# Производим проверку на важные признаки, если заранее созданный список 
# с важными признаками от SequentialFeatureSelection не был объявлен для ускорения выполнения

if ('IMP_SWS_RAW' not in locals()) or (IMP_SWS_RAW is None):
    imp_sfs_raw = list()
    for model_, X_, y_ in [[ridge, X, y],
                           [logr, X, y]]:
        list_forward, list_backward = sfs_research(model_, feature_names, X_, y_)
        imp_sfs_raw += list(list_forward) + list(list_backward)
else:
    imp_sfs_raw = IMP_SWS_RAW
````

###  Получение метрики на обучающей выборке с подбором гиперпараметров

#### Линейные модели - исследование без подбора гиперпараметров

Объявим функцию для исследования линейных моделей:

- LogisticRegression
- RidgeClassifier
- NN (tf.Keras Sequential)

Для вышеуказанных моделей гиперпараметры не подбираем, оценку производим кросс-валидацией с контролем балансировки классов, применяемых моделью самостоятельно. 

Функция будет создавать пайплайн процесса предобработки признаков и обучения модели для корректной проверки кросс-валидацией.

````python
def cv_research_linear(
    model_name_, model_,
    preprocessor_, X_, y_,
    scoring_=SCORING, cv_=CV_VALUE, verbose_=VERBOSE,
    n_jobs_=N_JOBS):
    '''
    '''
    pipe_ = make_pipeline(preprocessor_, model_)
    scores_ = cross_val_score(pipe_, X_, y_,
                              scoring=scoring_, cv=cv_, verbose=verbose_,
                              n_jobs=n_jobs_)
    return (abs(scores_)).mean()   
````

##### Модель NeuralNetwork - простая полносвязная нейронная сеть 

Построим и обучим простую нейронную сеть с тремя скрытыми полносвязными слоями слоями. 

Активатор выходного нейрона - логистическая регрессия.

В этой модели балансировку классов не применяем.

Для использования методов кросс-валидации от sklearn обернем созданную модель Sequential при помощи KerasClassifier от Scikeras для  использования в cross_vas_score.

````python
X_nn = features_train_one.to_numpy().astype(int)
y_nn = target_train.to_numpy().astype(int)

input_dim_ = X_nn.shape[1]
````

````python
def create_model():
    '''Функция создает простую модель NN Sequential используя TensorFlow.Keras
    Настройки сети невозможны, т.к. задаются имплементарно в теле функции
    Сеть состоит из трех скрытых полносвязных слоев в 20-2-8 нейронов, 
    без активатора, т.е. передаются только логиты нейронов.
    Выходной слой - 1 нейрон с активатором логистической регрессии
    Для работы функция требует объявления глобальных переменных 
    input_dim_ - int, входная размерность тензора с набором признаков объекта
    Возвращает Sequential.model
    
    '''
    
    model = Sequential()
    
    # Cлои 
    model.add(Dense(20, input_dim=input_dim_)) 
    model.add(Dense(2))  
    model.add(Dense(8))
    
    # Выходной слой
    model.add(Dense(1, activation='sigmoid'))
    
    # Сборка модели
    model.compile(optimizer=Adam(lr=0.01), 
                  loss='binary_crossentropy',
                  metrics=['AUC'])
    
    return model
````

````python
model_name = 'Neural_network'

model_nn = KerasClassifier(model=create_model, verbose=0, 
                           epochs=100,  
                           random_state=RANDOM_STATE)

nn_result = cv_research_linear(
    model_name, model_nn,
    preprocessor_scaler, X_nn, y_nn,
    cv_=3, scoring_=None, n_jobs_=N_JOBS)

print(f'\nМодель {model_name}. Средняя метрика при кросс валидации ROC-AUC={nn_result:.3}')
````

````python
class ThresholdExceeded(optuna.exceptions.OptunaError):
    pass

def check_threshold(study, trial):
    if study.best_value >= METRIC_GOOD_LIMIT:
        raise ThresholdExceeded()
    return
````

````python
N_EST_GB = (100, 500, 1)       # Список, количество итераций бустинга
ETA_GB = [0.005, 0.01, 0.05]            # Список, значения шага обучения
MAX_DEPTH_GB = (-1, 50, 1)  # Список, значения максимальной глубины дерева

N_EST_RF = (10, 110, 10)  # Диапазон количества решающих деревьев в случайном лесу, третье число - шаг
MAX_DEPTH_RF = (5, 50, 1)   # Диапазон количества макс глубины дерева в случайном лесу, третье число - шаг
MIN_SAMPLES_RF = (2, 3) # Диапазон значений мин. количества выборок, необходимых для разделения узла в случайном лесу
````

````python
# Вариант функции с базовым опредлелением обучения модели для RandomForestClassifier

def objective_rf(trial):    
    # Список гиперпараметров для перебора (для RandomForestClassifier)
    param = {
        'n_estimators': trial.suggest_int('n_estimators',
                                          N_EST_RF[0], N_EST_RF[1], N_EST_RF[2]),
        'max_depth': trial.suggest_int('max_depth', 
                                       MAX_DEPTH_RF[0], MAX_DEPTH_RF[1], MAX_DEPTH_RF[2]),
        'min_samples_split': trial.suggest_int('min_samples_split', 
                                               MIN_SAMPLES_RF[0], MIN_SAMPLES_RF[1])}
        
    estimator = RandomForestClassifier(
        **param, verbose=False, random_state=RANDOM_STATE,
        class_weight='balanced', n_jobs=N_JOBS)

    scores_ = cross_val_score(
        estimator, features_train_ordinal, target_train, 
        cv=CV_VALUE, scoring=SCORING)
    
    return (abs(scores_)).mean()
````

````python
%%time

model_name = 'RandomForestClassifier'

sampler_ = optuna.samplers.TPESampler()
pruner_ = optuna.pruners.HyperbandPruner()

# direction="maximize", т.к. чем больше метрика тем лучше
study = optuna.create_study(direction='maximize', 
                            sampler=sampler_, pruner=pruner_,
                            study_name='RF1')

try:
    study.optimize(objective_rf, n_trials=900, callbacks=[check_threshold])
except ThresholdExceeded:
    print(f'Выход по исключению, получено значение метрики:\
{study.best_value} >= {METRIC_GOOD_LIMIT}')

best_trial = study.best_trial
print(f'model: {model_name} tuned by Optuna, results:')
print(f'tuned param: {best_trial.params}')
print(f'ROC-AUC value on train set: {best_trial.value:.3}')
````
